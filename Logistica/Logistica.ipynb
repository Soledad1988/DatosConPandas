```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("Downloads/Transportation-Logistics-Tracking-Dataset_Spanish/Transportation & Logistics Tracking Dataset_Spanish.xlsx")
```

```python
#1️⃣ Distribución de los tipos de envío (Gráfico de barras)
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Tipo de Envío", hue="Tipo de Envío", palette="pastel", legend=False)

plt.title("Distribución de Tipos de Envío")
plt.ylabel("Cantidad")
plt.xticks(rotation=45)
plt.show()
```

```python
#2️⃣ Envíos a tiempo vs. tarde (Gráfico de barras)
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Tipo de Envío", color="pink")
plt.title("Envíos a Tiempo vs. Tarde")
plt.ylabel("Cantidad")
plt.show()
```

```python
#3️⃣ Distribución de la distancia de transporte (Histograma)
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Distancia de Transporte (KM)"].dropna(), bins=30, kde=True, color="skyblue")
plt.title("Distribución de la Distancia de Transporte")
plt.xlabel("Distancia (KM)")
plt.show()
```

```python
#4️⃣ Top 10 proveedores por cantidad de envíos (Gráfico de barras horizontal)
```

```python
top_proveedores = df["Nombre del Proveedor"].value_counts().head(10)

plt.figure(figsize=(8, 5))
sns.barplot(x=top_proveedores.values, y=top_proveedores.index, color="purple")
plt.title("Top 10 Proveedores por Cantidad de Envíos")
plt.xlabel("Cantidad de Envíos")
plt.show()
```
