## 4 - Limpieza de Datos

```python
df.fillna(0, inplace = True)
df
```

```python
#Corregir tipos de datos
df["Oro"] = df["Oro"].astype(int)
df["Plata"] = df["Plata"].astype(int)
df["Bronce"] = df["Bronce"].astype(int)
df
```

## Analisis de medallas por pais

```python
top_3_oro = df.sort_values('Oro', ascending = False).head(3)
top_3_oro
```

## Analisis de medallas Totales por pais

```python
filtro = df['Total'] > 10
mas_de_10_medallas = df[filtro]
mas_de_10_medallas.sort_values('Total', ascending = False)
```
