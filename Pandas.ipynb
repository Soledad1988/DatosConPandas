#DataFrame: Fundamentos y operaciones Bàsicas

```python
#El datFrame es una tabla con filas y columnas, donde las columnas son series de datos
```

```python
#Creamos un dataFrame:
data = [
    {'Nombre': 'Ana', 'Edad':25, 'Ciudad':'Madrid'},
    {'Nombre': 'Luis', 'Edad':30, 'Ciudad':'Barcelona'},
    {'Nombre': 'Pedro', 'Edad':22, 'Ciudad':'Venecia'}
  ]
df = pd.DataFrame(data)
df
```

```python
#Seleccionamos columnas
nombres = df['Nombre']
nombres
```

```python
#Añadir columnas
df['Salario'] = [30000, 40000, 25000]
df
```

```python
#borrar columna
df.drop('Salario', axis = 1, inplace=True)
```

```python
df
```

```python
#Seleccionamos la primera fila
primera_fila = df.iloc[0]
primera_fila
```

```python
#Selección con condición
df.loc[df['Nombre'] == 'Ana']
```

```python
df.loc[df['Edad'] >= 30]
```

```python
#Ordenar el dataFrame
df.sort_values(by='Edad', ascending = True) #ascending = False --> Descendente
```

#Series Temporales


```python
fechas = pd.date_range(start='2020-01-01', periods=6, freq='D')
valores = [34, 44, 65, 53, 39, 76]
serie_temporal = pd.Series(valores, index=fechas)
print(serie_temporal)
```

```python
print(serie_temporal['2020-01-03'])
```

```python
#Definimos diccionario
data = {
    'Fechas' : pd.date_range(start='2020-01-01', periods=6, freq='D'),
    'Ventas' : [345, 456, 567, 678, 234, 590]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Establecemos la columna Fechas como indice
df.set_index('Fechas', inplace=True)
print(df)
```
